Download a book from Gutenberg project in txt format.  Open in a text editor, delete all the pre-book text and post-book text (including the chapter list if there is one before the main text starts).
Make sure the word "Chapter" appears before all chapter breaks.  Because this is a pain, I did it for you (see Grimms.txt).

```python
# A utility function that mostly works on well-formatted Gutenberg text files with blank lines between chapter headings.

def ascii_print(x):
    try: 
        #return x.decode('utf8')
        return x.encode('ascii')
    except:
        return "*"
    
def read_chapters(filename): 
    """ This works on Gutenberg text files. Check for Chapter at start of line."""
    import re
    para = ''
    chapters = {}
    titles = {}
    count = 0
    chapters[count] = []
    titles[count] = []
    find_chapter = re.compile('^Chapter\s+', re.U|re.I)
    
    for line in open(filename):
        #print line
        line = line.decode('utf8').strip()
        if find_chapter.match(line):
            newtitle = line.replace('Chapter ','')
            print "matched ch", line, " count is ", count
            
            # dump all the saved paragraphs to the chapter dict
            if para.strip() != '' and ascii_print(para) != '*':   # in case of first chapter
                chapters[count].append(para)
                titles[count] = title
                para = ''
                count += 1
                chapters[count] = []
        if line.strip() != '' and ascii_print(line) != '*':  # filter out blank lines; remove \n from end of lines
            para = para + ' ' + line.strip()
            title = newtitle   # save the title you found, since you'll overwrite it with a new one
            
    # last one - no titles follow, so you need to dump this one!
    titles[count] = title
    chapters[count].append(para)
    return chapters, titles
```

### We need to write a csv file with the chapter text in a single cell, to use the single file option with the Topic Analysis GUI.

```python
def write_tsv(filename, id_dict, chapters_dict):
    """ Write out a CSV file with chapter titles in first col, chap in second."""
    with open(filename,'w') as output:
        for i, chap in chapters_dict.items():
            # wrap chapter in quotes because it contains commas
            #chap = ['"' + chap[0] + '"']
            output.write(id_dict[i] + '\t' + ' '.join(chap) + '\n')
```

```python
write_tsv("grimms_tsv.txt", titles, chaps)
```

```python
import csv
import collections

dir = 'topic_output/output_csv/'
topic_words = dir + 'Topics_Words.csv'
topic_docs = dir + 'TopicsInDocs.csv'
docs_topics = dir + 'DocsInTopics.csv'
```

```python
# pass in the topic_words filename
# This will give you the 10 words per topic; we asked for 10 topics
def list_words_for_topics(filename):
    words = []
    with open(filename, 'rb') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=',', quotechar='"')
        for index, row in enumerate(spamreader):
            if index > 0:
                words.append([x.strip() for x in row[1].split() if len(x.strip())>0])
    return words
```

```python
word_list = list_words_for_topics(topic_words)
len(word_list)  # you will have as many as topics you requested - so, 10.
```

```python
# pass in topic_docs.
# This gets you the topic assigments and strength per document.
def read_doctopics(filename):
    docs = []
    with open(filename, 'rb') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=',', quotechar='"')
        for row in spamreader:
            if spamreader.line_num > 1:   # skipping header row
                docid = row[0]
                topics = row[2:]
                topics_dict = dict(zip(topics[::2],topics[1::2]))
                print docid, topics_dict
                docs.append(collections.defaultdict(list))
                docs[-1] = topics_dict
    return docs
```

```python
docs_alltopics = read_doctopics(topic_docs)
print "Found docs ", len(docs_alltopics) # you should see the same as the number of docs, or chapters
```

### Heading into GEPHI, we need a more interesting format that includes nodes with edges with strengths...  We want documents and topics to be nodes, and relationships to be edges.  We want edge "thickness" to be proportional to the strength of the document association to a topic.

```python
# utility function for capturing output below and writing it to a file
def writeout(fname):
    with open(fname, 'w') as f:
        f.write(output.stdout)
```

```python
def remove_commas(instring):
    newstr = instring.replace(",", "")
    return newstr

# word list is the words per topic

def writegdfnodes(word_list, docs_alltopics, titles, documents):
    import operator
    print "nodedef>name VARCHAR, label VARCHAR, title VARCHAR, text VARCHAR"
    item = 0
    for docnum, doc in enumerate(docs_alltopics):
        item += 1  # nodes must start at 1, not 0!
        result = sorted(docs_alltopics[docnum].iteritems(), key=operator.itemgetter(1))
        print docnum + 1, ",", remove_commas(titles[docnum]), ",", remove_commas(titles[docnum]),  ",", "\"" + remove_commas(documents[docnum][0][0:360].encode('ascii')) + "...\""
    item += 1
    for topnum, topic in enumerate(word_list):
        print topnum + item, ",", " ".join(word_list[topnum][0:2]), ",", " ".join(word_list[topnum][0:2]), ",", "\"Top words:" + " ".join(word_list[topnum]) + "\""

# input docs_alltopics - this outputs the links, and weights
def writegdfedges(docs, weight_filter):
    offset = len(docs)
    # Gephi wants 'weight' and sigma.js wants 'size' - sigh.
    print "edgedef>node1 VARCHAR,node2 VARCHAR, weight DOUBLE, size DOUBLE"
    for docnum, doc in enumerate(docs):
        for key,val in doc.iteritems():
            if float(val) > weight_filter:
                print docnum + 1 , ",", int(key) + offset, ",", float(val) * 10.0, ",", float(val) * 10.0
```

```python
# After you run this, look at the output -- we used the top 2 words to label each topic node.
writegdfnodes(word_list, docs_alltopics, titles, chaps)
```

Now we run it with the capture setting... so we can save it to a file afterwards.

```python
%%capture output
writegdfnodes(word_list, docs_alltopics, titles, chaps)
```

```python
writeout('files/nodes.txt')   # we are saving the nodes we just created in the background to a file.
```

```python
# Notice that numbering for gephi starts at 1, not 0.
# Gephi uses 'weight' for edge thickness, but sigma.js wants 'size' for now.
writegdfedges(docs_alltopics, .15)
```

```python
%%capture output
writegdfedges(docs_alltopics, .15)  # we are filtering out weak edges.
```
